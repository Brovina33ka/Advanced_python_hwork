```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

### Задание 1

```python
x = np.linspace(1, 15,40)
def f(x):
    return np.sin(x/5)*np.exp(x/10) + 5*np.exp(-x/2)
plt.plot(x, f(x))
plt.show()
```

```python
t = np.array([1,8,15]) 
#свободный вектор b 
b = f(t)


#матрица А
A = np.array([[1, t[0], t[0]**2], [1, t[1], t[1]**2],[1, t[2], t[2]**2]])


#np.linalg.solve
X = np.linalg.solve(A, b)
print(X, 'ответ')

#многочлен
M2 = X[0] + X[1]*x + X[2]*x**2
```

```python
t = np.array([1,15])

#свободный вектор b 
b = f(t)


#матрица А
A = np.array([[1, t[0]], [1, t[1]]])


#np.linalg.solve
X = np.linalg.solve(A, b)
print(X, 'ответ')

#многочлен
M1 = X[0] + X[1]*x 
```

```python
t = np.array([1,4,10,15])

#свободный вектор b 
b = f(t)

#матрица А
A = np.array([[1, t[0], t[0]**2, t[0]**3], [1, t[1], t[1]**2, t[1]**3],[1, t[2], t[2]**2, t[2]**3], [1, t[3], t[3]**2, t[3]**3]])


#np.linalg.solve
X = np.linalg.solve(A, b)
print(X, 'ответ')

#многочлен
M3 = X[0] + X[1]*x + X[2]*x**2 + X[3]*x**3
```

```python
plt.plot(x, f(x),'red', label="точное решение")
plt.plot(x, M1,'blue', label="n = 2, 2 точки")
plt.plot(x, M2,'green', label="n = 2, 3 точки")
plt.plot(x, M3, 'black',label="n = 3, 4 точки")
plt.legend()
plt.show()
```

### Задание 4

В векторе x надо найти максимальный элемент среди тех, перед которыми стоит ноль.

```python
x = np.array([0, 11, 0, 0, -7, 2, 0, 4, 0])
#последний ноль не понадобится
x= np.delete(x, -1)
Zeroes = np.where((x == 0))[0] #найдем нули
max_kandidates = x[Zeroes+1]
print(np.max(max_kandidates))
```

### Задание 5

В векторе x надо заполнить каждый нулевой элемент предыдущим ненулевым значением

```python
y = np.array([8, 0, 0, 1, 0, 0, 0, -17.5, 0])
NotZeroes = np.arange(len(y)) #создаем массив такой же длины как y
NotZeroes[y == 0] = 0
print(NotZeroes)

NotZeroes = np.maximum.accumulate(NotZeroes)

Arr = y[NotZeroes]
print(Arr, 'ответ')
```
